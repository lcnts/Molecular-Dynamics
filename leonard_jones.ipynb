```python
import numpy as np
import matplotlib.pyplot as plt

# Lennard-Jones Parameters
epsilon = 2.0  # eV
sigma = 4.0    # Å

# Distance range
r = np.linspace(3.0, 10.0, 500)  # Å

# Lennard-Jones Potential Function
def lj_potential(r, epsilon, sigma):
    return 4 * epsilon * ((sigma / r) ** 12 - (sigma / r) ** 6)

# Compute potential energy
V = lj_potential(r, epsilon, sigma)

# Plot the potential energy curve
plt.figure(figsize=(8, 6))
plt.plot(r, V, label='Lennard-Jones Potential')
plt.axvline(x=2**(1/6) * sigma, color='r', linestyle='--', label='Energy Minimum')
plt.xlabel('Distance r (Å)')
plt.ylabel('Potential Energy V(r) (eV)')
plt.title('Lennard-Jones Potential Energy Curve')
plt.legend()
plt.grid(True)
plt.show()
```
